```python
A2 = cv.imread('A2.jpg')
plt.imshow(A2)
plt.show()
```

```python
img2 = cv.imread('A2.jpg',0)
```

```python
equ = cv.equalizeHist(img2)
res = np.hstack((img2,equ)) #stacking images side-by-side
cv.imwrite('res2.png',res)
```

```python
resimg2 = cv.imread('res2.png')
plt.imshow(resimg2)
plt.show()
```

# Observations

```python
import numpy as np

plt.subplot(1, 2, 1)
plt.hist(img1.flatten(),256,[0,256], color = 'r')
plt.title("Original A1 image")
plt.xlim([0,256])

plt.subplot(1, 2, 2)
plt.hist(resimg1.flatten(),256,[0,256], color = 'b')
plt.title("Equalized A1 image")
plt.xlim([0,256])

plt.tight_layout()
plt.show()
```

The histogram lies in brighter region, so to get the full spectrum the equalizeHist function maps the input pixels in brighter region to output pixels in full region.



```python
import numpy as np

plt.subplot(1, 2, 1)
plt.hist(img2.flatten(),256,[0,256], color = 'r')
plt.title("Original A2 image")
plt.xlim([0,256])

plt.subplot(1, 2, 2)
plt.hist(resimg2.flatten(),256,[0,256], color = 'b')
plt.title("Equalized A2 image")
plt.xlim([0,256])

plt.tight_layout()
plt.show()
```

In case of image A2, the histogram lies in the darker region and these darker pixels are mapped to the rest of the
 brighter region. Even though A2 is a darker image of A1, the equalized image is almost the same as the equalized image of A1 because equalization spreads the histogram evenly over the entire region for both images.
